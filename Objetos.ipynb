```python
class Saludador:
    def __init__(self, msg = "Hola Mundo "):
        self.__msg__ = msg
        
    def saludar(self, nombre="alejo"):
        print(self.__msg__ + nombre)
    
s = Saludador("Hola ")

s.saludar("maria")
print(s.__msg__)
```

```python
class UnidadTiempo:
    def __init__(self):
        self.__valor__ = 0
        self.__tope__ = 59
        
    def avanzar(self):
        if self.__valor__ < self.__tope__:
            self.__valor__ += 1
        else:
            self.__valor__ = 0
            
    def resetear(self):
        self.__valor__ = 0
        
    def obtener_valor(self):
        return self.__valor__
    
    def iniciar_valor(self, valor):
        self.__valor__ = valor

class Hora(UnidadTiempo):
    def __init__(self):
        self.__valor__ = 0
        self.__tope__ = 23
        
class Minuto(UnidadTiempo):
    def __init__(self):
        self.__valor__ = 0
        self.__tope__ = 59
        
class Segundo(UnidadTiempo):
    def __init__(self):
        self.__valor__ = 0
        self.__tope__ = 59
        
class Cronometro:
    def __init__(self):
        self.__h__ = Hora()
        self.__m__ = Minuto()
        self.__s__ = Segundo()
        
    def avanzar(self):
        self.__s__.avanzar()
        if self.__s__.obtener_valor() == 0:
            self.__m__.avanzar()
            if self.__m__.obtener_valor() == 0:
                self.__h__.avanzar()
                
    def iniciar_tiempo(self, valor):
        lista = [int(x) for x in valor.split(":")]
        self.__h__.iniciar_valor(lista[0])
        self.__m__.iniciar_valor(lista[1])
        self.__s__.iniciar_valor(lista[2])
        
        
    def obtener_tiempo(self):
        return "{:02}:{:02}:{:02}".format(self.__h__.obtener_valor(), self.__m__.obtener_valor(), self.__s__.obtener_valor())
    

c = Cronometro()
c.iniciar_tiempo("01:10:30")


for i in range(62):
    c.avanzar()
    print(c.obtener_tiempo())
```

```python
from math import sqrt, pi

class Punto:
    def __init__(self, x = 0, y = 0):
        self.__x__ = x
        self.__y__ = y
        
    def calcular_distancia(self, punto):
        return sqrt((self.__x__ - punto.__x__)**2 + (self.__y__ - punto.__y__)**2)
    
class Figura:
    def __init__(self, p1 = Punto(), p2 = Punto()):
        self.__origen__ = p1
        self.__fin__ = p2
        
    def calcular_area(self):
        pass

class Circulo(Figura):
    def calcular_area(self):
        radio = self.__origen__.calcular_distancia(self.__fin__)
        area = pi * radio ** 2
        return area

class Cuadrado(Figura):
    def calcular_area(self):
        lado = self.__origen__.calcular_distancia(self.__fin__)
        area = lado ** 2
        return area

class Rectangulo(Figura):
    def calcular_area(self):
        punto_temp = Punto(self.__origen__.__x__, self.__fin__.__y__)
        base = punto_temp.calcular_distancia(self.__origen__)
        altura = punto_temp.calcular_distancia(self.__fin__)
        area = base * altura
        return area
    
class TrianguloRectangulo(Figura):
    def calcular_area(self):
        punto_temp = Punto(self.__origen__.__x__, self.__fin__.__y__)
        base = punto_temp.calcular_distancia(self.__origen__)
        altura = punto_temp.calcular_distancia(self.__fin__)
        area = base * altura / 2
        return area
    
f = Cuadrado(Punto(0, 0), Punto(0, 5))
print(f.calcular_area())
```
